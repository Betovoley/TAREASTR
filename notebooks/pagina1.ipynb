```python
import os
import pandas as pd
from pathlib import Path
from datetime import datetime
```

```python
# Cambia 'ruta_del_archivo.csv' por la ruta de tu archivo
ruta_archivo = 'C:\\Users\\Betovoley\\Videos\\Master en Python 1 edicion\\Modulo 8\\TAREASTR\\data\\all_games_2425.csv'
df = pd.read_csv(ruta_archivo)

# Mostrar las primeras filas del DataFrame
df
```

```python
# Filtrar las filas donde 'skill' sea igual a 'serve'
df_filtrado = df[df['skill'] == 'Serve']

# Seleccionar las columnas que quieres guardar en el nuevo CSV
columnas_a_guardar = ['time', 'player_name', 'player_role', 'skill', 'evaluation_code', 'start_zone', 'end_zone', 'end_subzone', 'set_number',
                       'team_id']  # Reemplaza con los nombres de las columnas que necesitas
df_filtrado = df_filtrado[columnas_a_guardar]

# Exportar a un nuevo archivo CSV
df_filtrado.to_csv('saque_limpio.csv', index=False)

print("Archivo CSV creado con éxito.")
```

```python
df_filtrado.isna().sum()
```

```python
# Tratamiento de los valores nulos y vacios.
df_filtrado['start_zone'] = df_filtrado['start_zone'].fillna(0)
df_filtrado['end_zone'] = df_filtrado['end_zone'].fillna(0)

df_filtrado.to_csv('saque_limpio.csv', index=False)

print("Archivo CSV guardado con exito")
```

```python
# Convertir a datetime manejando errores
df_filtrado['time'] = pd.to_datetime(df['time'], errors='coerce')

# Extraer solo la fecha, ignorando los valores no válidos (NaT)
df_filtrado['time'] = df_filtrado['time'].dt.date

df_filtrado.to_csv('saque_limpio.csv', index=False)

print("Archivo CSV guardado como exito")

df_filtrado
```
